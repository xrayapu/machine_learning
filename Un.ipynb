```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as mp
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
x = DataFrame(data,columns=['production_budget_usd'])
y = DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
mp.figure(figsize=(10,6))
mp.scatter(x,y,alpha=0.3)
mp.title('cost vs revenue')
mp.xlabel('cost')
mp.ylabel('revenue')
mp.ylim(0,3e9)
mp.xlim(0,4.5e8)

mp.show()
```

```python
reg=LinearRegression()
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
mp.figure(figsize=(10,5))
mp.scatter(x,y,alpha=0.3)
mp.title('cost vs revenue')
mp.xlabel('cost')
mp.ylabel('revenue')
mp.ylim(0,3e9)
mp.xlim(0,4.5e8)
mp.plot(x,reg.predict(x), color='gray', linewidth=3)

mp.show()
```

```python
reg.score(x,y)
```
